```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense,Dropout,BatchNormalization
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
import tensorflow as tf
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
tf.data.experimental.enable_debug_mode()
```

```python
df_input=pd.read_csv("MailOutput_lat.csv")
df_output=pd.read_csv("Clustered_emails.csv")
df_output
```

```python
df_combined=pd.merge(df_input,df_output,left_on="MessageID",right_on="ID").drop(['Unnamed: 0', 'ID'], axis=1)
df_combined.set_index("MessageID",inplace=True)
df_combined=df_combined.replace({'Status': {'Read': True, 'Unread': False}})
df_combined
df_combined.to_csv('MailOutput_combined.csv', index=False, encoding="utf-8-sig")
```

```python
df_X = df_combined.loc[:, 'EmailSize':'isBCC'].to_numpy(dtype ='float32')
df_y = df_combined.loc[:, 'Topic'].to_numpy()
df_Xtrain, df_Xtest, df_Ytrain, df_Ytest = train_test_split(df_X, df_y, test_size=0.2)
df_YtrainArray = to_categorical(df_Ytrain,num_classes=5,dtype ="int32")
df_Xtrain.size
```

```python
model = Sequential()
model.add(Dense(512, activation='relu', input_dim=12))
model.add(Dropout(0.3))
model.add(Dense(256, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(128, activation='relu'))
model.add(Dense(5, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(df_Xtrain,df_YtrainArray,validation_split = 0.1, epochs=1000, batch_size=4)
```

```python
model.evaluate(df_Xtrain,df_YtrainArray)
```

```python
output=model.predict(df_Xtest)
y_pred = tf.argmax(output, axis=1)
output
```
